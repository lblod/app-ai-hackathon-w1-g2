```python
import cv2
import numpy as np
from pdf2image import convert_from_path
from google.colab.patches import cv2_imshow
import pytesseract
import time
```

#Tesseract

```python
def ocr_tess (pdf_path):
  t1 = time.time()
  pages = convert_from_path(pdf_path, 500)
  output_file_tess = 'ocr_output_tess.txt'
  with open(output_file_tess, 'w') as f:
    for count, page in enumerate(pages):
        #page.save(f'images/out{count}.jpg', 'JPEG')
        img_cv = np.array(page)
        img_gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
        #cv2_imshow(img_gray)
        _, thresh = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY)
        #cv2.imwrite(f'images/gray{count}.jpg', img_gray)
        text = pytesseract.image_to_string(thresh)
        # Print or process the extracted text
        f.write(f"Text from page {count + 1}:\n")
        f.write(text)
        f.write("\n" + "-" * 80 + "\n")
  t2 = time.time()
  tt = t2-t1
  print(tt, "seconds")
```

```python
ocr_tess ('23878.pdf')
```

```python
from PIL import Image
from surya.ocr import run_ocr
from surya.model.detection.model import load_model as load_det_model, load_processor as load_det_processor
from surya.model.recognition.model import load_model as load_rec_model
from surya.model.recognition.processor import load_processor as load_rec_processor
```

```python
def ocr_surya (pdf_path):
  t1 = time.time()
  pages = convert_from_path(pdf_path, 500)
  output_file_surya = 'ocr_output_surya.txt'
  with open(output_file_surya, 'w') as f:
    for count, page in enumerate(pages):
        langs = ["en"] # Replace with your languages - optional but recommended
        det_processor, det_model = load_det_processor(), load_det_model()
        rec_model, rec_processor = load_rec_model(), load_rec_processor()
        predictions = run_ocr([page], [langs], det_model, det_processor, rec_model, rec_processor)
        extracted_text = []
        for ocr_result in predictions:
            for text_line in ocr_result.text_lines:
                extracted_text.append(text_line.text)
        # Print or process the extracted text
        f.write(f"Text from page {count + 1}:\n")
        for text in extracted_text:
            f.write(text)
        f.write("\n" + "-" * 80 + "\n")
  t2 = time.time()
  tt = t2-t1
  print(tt, "seconds")
```

```python
ocr_surya ('23878.pdf')
```
